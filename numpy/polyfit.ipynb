```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt

path = r'D:\Desktop\高濃度晶纖'
os.chdir(path)
os.listdir()
```

```python
file = r'裸光纖量測資料.xlsx'
data = pd.read_excel(file)
x = data.index
y = data['ASE']

plt.plot(x, y, 'ro')
plt.show()
```

```python
fit = np.polyfit(x ,y ,2)
print(fit)
```

```python
eq = np.poly1d(fit)
s = np.linspace(0, 2000, 51)
plt.plot(s, eq(s), x, y, 'ro')
```
