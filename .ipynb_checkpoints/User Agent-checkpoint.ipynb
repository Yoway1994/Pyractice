```python
User Agent的意思是指誰來代替我們訪問這個網站，而 request header則是一段請求時所夾帶的信息，

可以用網頁檢查時，點選network，可以在header下面看到這段訊息，



但是當我們用爬蟲訪問時，通常UA是python的request庫，我在這裡安裝一個 requests 模組來看看header

import requests

r = requests.get('https://www.ptt.cc/bbs/Beauty/index.html')
print(r.request.headers)

>>>
{'User-Agent': 'python-requests/2.20.1', 'Accept-Encoding': 'gzip, deflate', 'Accept': '*/*', 'Connection': 'keep-alive'}

可以看到當我們用python爬蟲去訪問網站時，用的是這個UA身分。

所以對於反爬蟲的網站，我們只要以其他瀏覽器的身分就可以成功得到網站訊息。

https://developer.mozilla.org/zh-TW/docs/Web/HTTP/Headers/User-Agent

可以到上面的網站去看看其他瀏覽器或其他UA。


```
