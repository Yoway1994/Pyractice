```python
2018/11/05
在用python做爬蟲工具時，Regular Expression，正則表達式為一相當好用的工具，

from urllib.request import urlopen
import re

p1 = 'Beauty'
html = urlopen('https://www.ptt.cc/bbs/Beauty/index.html').read().decode('utf-8')

print(p1 in html)

>>> True

我們在urllib 模組使用筆記的一開始讀取了一個網站的頁源碼，在這裡我用ptt表特版做為範例，

我設定p1 = 'Beauty' 做為關鍵字，接著利用 'p1 in html' 去檢驗這個關鍵字有沒有在該被讀取的頁源碼中。

可以看到結果為 True，表示存在。

我們嘗試去搜尋這個關鍵字的位置，利用re.search這個語法

print(re.search(p1, html))

>>> <re.Match object; span=(140, 146), match='Beauty'>

你可以看到他返回的結果告訴我們，在從第140~146的字元找到了這個關鍵字。

我們進一步強化他的搜尋功能，

p1 = '[Bb]eauty'
print(re.search(p1, html))

假設我們今天不知道要搜尋的對象是大寫或小寫，或是中間有替換字元，我們可以使用中括號

以上面的例子就是 'beauty' 和 'Beauty' 都可以被搜尋到。

2018/11/09
我將一些正則表達式的符號整理在re.search中，在這裡會慢慢增加其他表達式的符號。

我們有使用過問號 '?' 當作 may or may not 表示法為 ( )?，也同意於{0,1}

但如果放在一些限制符號後面，則表示，盡可能匹配最少字符，

import re

string = 'woooooooooo'
p1 = 'wo*'
p2 = 'wo*?'

print(re.search(p1, string))
print(re.search(p2, string))

>>>
<re.Match object; span=(0, 11), match='woooooooooo'>
<re.Match object; span=(0, 1), match='w'>

如果是沒有問號，正則表達會匹配字符最多的可能，加了問號則是匹配最少。
```
