```python
相較於僅顯示True or False這種存在與否的表示，我們可以利用search去找出位置，

並且可以很針對要找得對象是甚麼，藉此進一步強化他的搜尋功能。

首先我們可以單純尋找字串存在的位置，

string = 'fine me here'
p1 = 'here'
print(re.search(p1, string))

>>> 
<re.Match object; span=(8, 12), match='here'>

假設我們今天不知道要搜尋的對象是大寫或小寫，或是中間有替換字元，我們可以使用中括號，

像是對於一些動詞的變化型可以用如此方法，像是 'w[io]n'

就是表示對 'win' 或者 'won' 都可以搜尋到該關鍵字。

str1 = 'I win the game'
str2 = 'yesterday, I won the game'
p1 = 'w[io]n'

print(re.search(p1, str1))
print(re.search(p1, str2))

>>> 
<re.Match object; span=(2, 5), match='win'>
<re.Match object; span=(13, 16), match='won'>

如果要亂數隨機亂找，我們甚至可以使用'[A-Z]' 或 '[a-z]' ，'[0-9]' 這種對全部符號進行搜尋

也可以組合，'[a-z0-9A-Z]' 便可以帶入任意符號搜尋。

str = 'find something here.'
find = '[A-Za-z0-9]ind'

print(re.search(find, str))

>>> 
<re.Match object; span=(0, 4), match='find'>

它自動幫我們match到了關鍵字 'find'

但這種打法稍嫌麻煩，所以我們用了 '\d' 和 '\D' 來代表 "全體數字" 和 "非全體數字"

str = 'find something here.'
find = '\Dind'

print(re.search(find, str))

>>> 
<re.Match object; span=(0, 4), match='find'>

或者更簡單我們使用 '\w' (word)

str = 'find something here.'
find = '\win\w'

print(re.search(find, str))

>>>
<re.Match object; span=(0, 4), match='find'>

你可以看到 '\w' 不僅包含所有大小寫英文字母，數字，且連底線 '_' 也一並包含。

'\s' 就是代表'space'，在文本編輯中任何空格，或回車轉義方法，'\S' 大寫一樣則一樣則為其相反。



接著是搜尋單詞的'\b'，我們先看這個例子

str = 'find something here.'
p1 = 'some'

print(re.search(p1, str))

>>> 
<re.Match object; span=(5, 9), match='some'>

當我們心裡只想匹配 'some' 這個單獨詞彙，但卻因為 'something' 裡面也含有 'some'，

所以它也被誤認匹配到 'some' 這個單字，但實際該句子卻沒有。

所以我們得用'\b'來指定該單字，

str = 'find something here.'

p1 = r'\bsome\b'
p2 = r'\bsomething\b'

print(re.search(p1, str))
print(re.search(p2, str))
 
>>> 
None
<re.Match object; span=(5, 14), match='something'>

向第一個p1，雖然 'something' 裡面包含 'some' ，

但因為我們指定, '\bsome\b'，some必須為一前後有空格的單一單字才能被找出來，故結果是None

同理 '\B' 就是反過來，如果包含了空格就不會被找出來。

str = 'find something here.'

p1 = r'\Bsome\B'
p2 = r'some\B'

print(re.search(p1, str))
print(re.search(p2, str))

>>> 
None
<re.Match object; span=(5, 9), match='some'>

你可以看到 '\B' 就是排除了 '\b' 以外的所有case。

在這裡我們要注意一點就是，我在帶有'\b'的字串前面加上一個 r，

因為'\b'本身在python已經有轉義字符了，必須加上 r 否則得再用雙斜槓來表示這個效果。

p1 = '\\bsomething\\b'

再甚至你可以匹配任何一個字符，包含任何一個符號或空格，但不包跳行

string = 'Would you want to continue [YES/NO]?'
p1 = '.YE..N..'
print(re.search(p1, string))

>>> 
<re.Match object; span=(23, 31), match='[YES/NO]'>

你可以看到他不只匹配字母，連一些框格或符號都可以匹配。

接著我們可以指定要匹配的單字是否在字首或字尾，

string = 'The first one.'

p1 = "^The"
p2 = ".$"

print(re.search(p1, string))
print(re.search(p2, string))

>>> 
<re.Match object; span=(0, 3), match='The'>
<re.Match object; span=(13, 14), match='.'>

有時候會遇到一種情況是，該單字明明在句首，但卻無法找到，這時候我們要增加一個multiline的變數，

string = '''
Today is Saturday.
I don't want to go outside.
'''
p1 = "^I"

print(re.search(p1, string))
print(re.search(p1, string, flags = re.M))

>>> 
None
<re.Match object; span=(20, 21), match='I'>

如此一來可以逐行逐行下去尋找。

有個情況是你不確定單字有沒有被縮寫的時候也可以匹配，

str1 = 'Today is Friday.'
str2 = 'Today is Fri.'

p1 = "Fri(day)?"

print(re.search(p1, str1))
print(re.search(p1, str2))

>>> 
<re.Match object; span=(9, 15), match='Friday'>
<re.Match object; span=(9, 12), match='Fri'>

就是說我不清楚，括號裡面的東西有沒有的時候可以使用

在來是重複匹配0次或多次以上，

str1 = 'you are wy too far.'
str2 = 'you are waaaaaaaaaaaaaaaaaaay too far.'
p1 = 'a*'

print(re.search(p1, str1))
print(re.search(p1, str2))

>>>
<re.Match object; span=(8, 10), match='wy'>
<re.Match object; span=(8, 29), match='waaaaaaaaaaaaaaaaaaay'>

也有限制一次以上的，換成加號就可以了

str1 = 'yes I have eyes.'
str2 = 'eeeeeeeyes'
p1 = 'e+yes'

print(re.search(p1, str1))
print(re.search(p1, str2))

>>>
<re.Match object; span=(11, 15), match='eyes'>
<re.Match object; span=(0, 10), match='eeeeeeeyes'>

當然也可以限定次數出現的範圍，或只表示特定次數

string = 'n no noo nooooooooo'
p1 = 'no{2,5}'
p2 = 'no{9}'
print(re.search(p1, string))
print(re.search(p2, string))

>>>
<re.Match object; span=(5, 8), match='noo'>
<re.Match object; span=(9, 19), match='nooooooooo'>

最後來張整理

[ ]
certain index in [ ]
\d
\D
decimal digit
any non-decimal digit
\w
\W
[a-zA-Z0-9_]
opposite to \w
\s
\S
any white space [\t\n\r\f\v]
opposite to \s, any non-white space
\b
\B
empty string (only at the start or the end of  the word)
empty string (but not at the start or the end of  the word)
.
\\
match anything (except \n)
^
$
match line beginning
match line ending
( )?
may or maynot occur = {0,1}
*
+
occur 0 or more times
occur 1 or more times
{n,m}
{n}
occur n to m times
only occur n times



```
