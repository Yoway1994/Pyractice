```python
2018/11/01
為了嘗試模擬laser rate equation，其中必須使用到微分方程所以才嘗試使用這個模組。

此模組非內建，需先用pip進行安裝。

一開始我們先將所有語法載入進來

from sympy import * 

相較於其他import法，這方法可以省去我在每個語法前打sy. 像是

import sympy as sy
sy.diff()
sy.sin()
sy.cos()
.
.
.

或者你可以逐行增加你所需要的語法，以避免import太多東西

from sympy import diff
from sympy import sin
.
.
.


sympy一開始遇到特點是，所有你要用的變數都必須先宣告

from sympy import *
x = Symbol('x')
y = Symbol('y')

經宣告後的變數可以直接進行運算，不須先行帶入數值，如此以來可以直接表達一個多項式。

from sympy import *
x = Symbol('x')
y = Symbol('y')

print(2*x+y-3)

>>> 2*x + y -3

接著我們嘗試微分這個語法，首先我利用sympy.diff將一個多項式微分之後並利用sympy.plot畫出來

from sympy import *
x = Symbol('x')
y = Symbol('y')

y = x**3 + 1
plot(y.diff(x), (x,0,1))

 
微分後的方程如圖所示，且我將x範圍設在(0, 1)之間。

2018/11/02
接著我們要開始來考慮微分方程，要解微分方程得用到sympy.dsolve這個語法

from sympy import *
x = Symbol('x')
f = Function('f')

eq_solve = dsolve(f(x).diff(x) - f(x))

print(eq_solve)

>>> Eq(f(x), C1*exp(x))

我們在這裡宣告一個新的class變數叫做 'Function' ，你可以在指定一個變數在他方程式裡，像是f(x),

假設我們今天要解的方程式是 f(x) = diff(f(x), x)，也就是f(x)的微分等於自己，

我們可以移動各項讓其合為零，像是 diff(f(x), x) - f(x) = 0

如此便可以用dsolve語法去解這個方程式，再將這個結果assign給一個變數就行了。

解出來的方程式因為沒有設邊界條件，所以會有一個常數C1，

這時可以在dsolve內增加一個ics變數，這樣C1就可以解出一特定值，

eq_solve = dsolve(f(x).diff(x) - f(x), ics = {f(0): 1})

print(eq_solve)

>>> Eq(f(x), exp(x))

在這裡我將 f(0) = 1 代入使其成為條件，並解出C1 = 1 

當然boundary condition也可以是一個待輸入的變數，相當適合用在物理上作為常數使用，

G = Symbol('G')
ics = {f(x): G}

eq_solve = dsolve(f(x).diff(x) - f(x), ics = ics)

print(eq_solve)

>>> Eq(f(x), G*exp(x))

在給定邊界條件後，最後我們把解出來的方程式畫出來。

我們再shell上輸出的結果是，Eq(f(x), exp(x))，他的意思是其解為 f(x) = exp(x)

我們使用一個語法 .rhs (right hand side)，去得到等式右邊的解，最後就可以用plot畫出來

eq_solve = dsolve(f(x).diff(x) - f(x), ics = {f(0): 1})
plot(eq_solve.rhs, (x, 0, 1))

同理如果輸出eq.lhs (left hand side) 就會得到 f(x)。

```
