```python
2018/09/15
csv模組為內建標準模組，不需先行安裝。

import csv

注意不要把你的檔名取作csv.py，如果名稱跟模組相同，它會導入自身這個py檔而不是csv模組。

導入之後接著讀取之前catch的OSA頻譜csv檔

file_name = "spectrum_name.csv"

with open(file_name,'r') as csv_file:
    csv_reader = csv.reader(csv_file)

相較於直接使用.open的方法，這裡配合使用with 陳述句可以省去使用.close以避免忘記。

接著我們開始讀取裡面的檔案，

with open(file_name,'r') as csv_file:
    csv_reader = csv.reader(csv_file)
    
    for line in csv_reader:
        print(line)

因為在這裡csv_reader並不是一個 list，如果直接試著輸出reader，

print(csv_reader)

>>>

<_csv.reader object at 0x010E3CF0>

只會顯示該物件被儲存在哪個記憶體位置，所以我們才用"for line in file"去把每一行的內容讀取出來。

2018/09/18
在正式取用數據前，csv檔前面數行有我們不需要的文字，我們可以用enumerate語法去選取我們需要的行數段落。

for i, rows in enumerate(csv_reader):
    if i > 28:
        print(rows)

你也可以擷取特定一行,

if i == 2
    print(rows)

選定資料範圍之後，我們將它加到一個list裡面

wav = []
val = []

for i, rows in enumerate(csv_reader):
    if i > 28:
        wav.append(rows[0])
        val.append(rows[1])

於是我們就有了兩個list，在這裡我是用一張頻譜作範例，所以wav代表波長而val代表強度。

再來用wav當x座標, val當y座標，就可以將這張頻譜圖用matpltlib畫出來。

2018/09/21
如果想知道文件的行數，有個簡單的辦法可以檢驗

count = 0

for i in csv_reader:
    count += 1

print(count)

這方法比起"len"函數運算來得快一點。


2018/09/27
在座圖之前我想把抓出來的資料做些運算，在這個例子中我把mW轉換成dBm

from numpy import log10

val = []
wav = []

for i, rows in enumerate(csv_reader):
    if i > 28:
        v = 10 * log10(abs(float(rows[1])))
        val.append(v)
        wav.append(rows[0])

在這裡從'numpy'裡面叫出'log10'來進行運算,並使用abs將我的資料先取絕對值。

使用'float'語法是因為，原本存在list中的資料型態為'string',需先轉換成浮點小數才能做後續數學運算。

最後再加上matplotlib整個程式碼看起來像這樣。

import csv 
from numpy import log10
import matplotlib.pyplot as plt

file_name = '000.csv'
with open(file_name, 'r') as csv_file
    csv_reader = csv.reader(csv_file)
    
    val = []
    wav = []
    
    for i, rows in enumerate(csv_reader):
        if i > 28:
            v = 10 * log10(abs(float(rows[1])))
            val.append(v)
            wav.append(rows[0])

   plt.plot(wav,val)
   plt.show()


這張頻譜圖就是最後的成果。

2018/12/18
當我們完成一張頻譜圖，接著我們要處理多筆頻譜資料，在os 模組使用筆記中我們已經找到檔案的位置，

接著我們用for迴圈把他們逐一開啟並處理資料。

for file in os.listdir():
    file_path = os.path.join(path, file)
    with open(file_path, 'r') as csv_file:
        csv_reader = csv.reader(csv_file)

那開啟後資料為了不讓for迴圈過於龐大，我把數據處理的部分寫成兩個function

def pure_val(csv_reader):
    list = []
    for i, rows in enumerate(csv_reader):
        if i > 28:
            np.seterr(divide = 'ignore')
            v = 10 * np.log10(abs(float(rows[1])))
            list.append(v)
    return list
    
def pure_wav(csv_reader):
    list = []
    for i, rows in enumerate(csv_reader):
        if i > 28:
            list.append(round(float(rows[0]),1))
    return list

這裡pure_val 就是讓我得到dBm，然後pure_wav就是得到對應的波長。

這兩個function基本上，就是抓出數字並放進list返回。

total_x = []
total_y = []    

for file in os.listdir():
    file_path = os.path.join(path, file)
    with open(file_path, 'r') as csv_file:
        csv_reader = csv.reader(csv_file)
        total_x.append(pure_wav(csv_reader))

for file in os.listdir():
    file_path = os.path.join(path, file)
    with open(file_path, 'r') as csv_file:
        csv_reader = csv.reader(csv_file)
        total_y.append(pure_val(csv_reader))

最後再把每一組list 又存到一個 total_list裡面，

像這樣total_x[0] 和 total_y[0]就是第一份csv的頻譜資料，最後再用matplotlib畫出來。

import csv
import numpy as np
import os
import matplotlib.pyplot as plt

path = r'D:\DeskTop\tuning\1800mW'
os.chdir(path)

def pure_val(csv_reader):
    list = []
    for i, rows in enumerate(csv_reader):
        if i > 28:
            np.seterr(divide = 'ignore')
            v = 10 * np.log10(abs(float(rows[1])))
            list.append(v)
    return list
    
def pure_wav(csv_reader):
    list = []
    for i, rows in enumerate(csv_reader):
        if i > 28:
            list.append(round(float(rows[0]),1))
    return list

total_x = []
total_y = []    

for file in os.listdir():
    file_path = os.path.join(path, file)
    with open(file_path, 'r') as csv_file:
        csv_reader = csv.reader(csv_file)
        total_x.append(pure_wav(csv_reader))


for file in os.listdir():
    file_path = os.path.join(path, file)
    with open(file_path, 'r') as csv_file:
        csv_reader = csv.reader(csv_file)
        total_y.append(pure_val(csv_reader))


for i in range(len(total_y)):
    plt.plot(total_x[i], total_y[i], lw = 0.5)

plt.ylim((-55, -10))    
plt.xlim((1350, 1500))
plt.xticks(np.arange(1370, 1500, 20))
plt.show()


```
